```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python

# Scroll the page to load dynamic content
def scroll_page(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
```

```python

# Click 'Load More Trucks' button until all trucks are loaded
def load_all_trucks(driver):
    while True:
        try:
            # Find the "Load More Trucks" button
            load_more_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, "//span[contains(text(),'Load More Trucks')]"))
            )
            # Scroll to the button and click it using JavaScript to avoid interception issues
            driver.execute_script("arguments[0].scrollIntoView(true);", load_more_button)
            driver.execute_script("arguments[0].click();", load_more_button)
            print("Clicked 'Load More Trucks' button.")
            time.sleep(5)  # Wait for more trucks to load
        except Exception as e:
            print(f"No more 'Load More Trucks' button found or error: {e}")
            break  # Exit the loop if the button is not found
```

```python

# Set up ChromeOptions to disable/handle permissions
chrome_options = Options()
chrome_options.add_argument("--disable-notifications")  # Disable notifications
chrome_options.add_argument("--disable-geolocation")    # Disable geolocation prompts

# Install and set up ChromeDriver using WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python

# Set an explicit wait to handle dynamic elements
wait = WebDriverWait(driver, 15)  # Wait up to 15 seconds for elements to appear
```

```python

# Open the main page containing all EV 3-wheelers
driver.get('https://trucks.cardekho.com/en/find/truck-3-wheeler+auto-rickshaw+electric')
print("Opened the main EV 3-wheeler page.")
```

```python

# Allow the page to load and for any permission popups to appear
time.sleep(5)

# Click the "Load More Trucks" button until all trucks are loaded
load_all_trucks(driver)
```

```python

# Collect all vehicle links (specifications links)
vehicle_links = []
vehicle_names = []

try:
    # Find the download buttons using XPath for the parent <a> tag with 'DownloadBroucher' class
    download_buttons = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//a[@class='DownloadBroucher']")))
    print(f"Found {len(download_buttons)} download buttons on the main page.")

    for button in download_buttons:
        link = button.get_attribute('href')
        
        # Find the vehicle name by checking the nearest heading or element above the download button
        try:
            parent_element = button.find_element(By.XPATH, "./../../..")  # Adjust parent to reflect the hierarchy correctly
            name = parent_element.find_element(By.XPATH, ".//h3").text  # Look for an <h3> element inside the parent
        except Exception as name_error:
            name = "Unknown Model"
            print(f"Error finding model name: {name_error}")
        
        print(f"Found brochure link: {link} | Model Name: {name}")
        vehicle_links.append(link)
        vehicle_names.append(name)
except Exception as e:
    print(f"Error finding download buttons: {e}")

# At this point, we have collected all the truck links
print(f"Total trucks collected: {len(vehicle_links)}")

# Save the truck links and names to a CSV
df = pd.DataFrame({
    'Truck Name': vehicle_names,
    'Specification Link': vehicle_links
})
#df.to_csv('collected_truck_links.csv', index=False)

# Close the browser session
driver.quit()

print("Truck links fetched and saved successfully!")
```

```python
# import re
# import pandas as pd
# from selenium import webdriver
# from selenium.webdriver.chrome.service import Service
# from webdriver_manager.chrome import ChromeDriverManager
# from selenium.webdriver.common.by import By
# from selenium.webdriver.chrome.options import Options
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC
# import time

# Extract the average price from the price string
def extract_average_price(price_string):
    price_numbers = re.findall(r'([0-9.]+)', price_string)
    if len(price_numbers) == 2:
        price_1 = float(price_numbers[0])
        price_2 = float(price_numbers[1])
        return (price_1 + price_2) / 2
    elif len(price_numbers) == 1:
        return float(price_numbers[0])
    return None

# Scroll the page to load dynamic content
def scroll_page(driver):
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)  # Reduced the sleep time for quicker scrolling
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

# Set up ChromeOptions to disable/handle permissions and run in headless mode
chrome_options = Options()
chrome_options.add_argument("--disable-notifications")
chrome_options.add_argument("--disable-geolocation")
chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1920,1080")

# Install and set up ChromeDriver using WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# Set an explicit wait to handle dynamic elements
wait = WebDriverWait(driver, 15)

# Load the truck links CSV file from Step 1
df_links = pd.read_csv('collected_truck_links.csv')

# Prepare to collect the specifications
specifications = []

for index, row in df_links.iterrows():
    serial_id = index + 1  # Serial ID starting from 1
    truck_name = row['Truck Name']
    truck_link = row['Specification Link']
    
    # Print the Serial ID once per truck
    print(f"Serial ID: {serial_id} - Visiting brochure page for {truck_name}")
    
    driver.get(truck_link)  # Visit the brochure page
    time.sleep(3)  # Shortened wait time to load the page

    try:
        # Get the specifications page link if available
        spec_link = wait.until(EC.element_to_be_clickable((By.XPATH, "//a[contains(text(),'Specifications')]")))
        spec_link.click()
        time.sleep(3)  # Shortened wait time after clicking
        
        # Scroll the page to load all dynamic specs sections
        scroll_page(driver)

        specs = {"Serial ID": serial_id, "Name": truck_name}

        # Extract all tables under different sections using "data-track-section"
        sections = ["Performance", "Charging", "Dimension", "Transmission & Loading Capacity", 
                    "Features", "Brakes & Suspension", "Body Option & Cabin Type", "Tyres", "Others"]
        
        for section in sections:
            try:
                # Locate the table associated with the section using the 'data-track-section'
                table = driver.find_element(By.XPATH, f"//table[@data-track-section='{section}']")
                rows = table.find_elements(By.XPATH, ".//tr")
                
                for row in rows:
                    try:
                        key = row.find_element(By.XPATH, "./td[1]").text
                        value = row.find_element(By.XPATH, "./td[2]").text
                        specs[f"{section} - {key}"] = value  # Store with section as part of the key
                    except:
                        pass  # Silently pass over errors to keep things quick
            except:
                pass  # Pass silently if a section is not found

        # Now fetch the price data
        try:
            price = driver.find_element(By.XPATH, "//div[contains(@class,'price')]").text
            average_price = extract_average_price(price)
            specs["Price (Lakh INR)"] = average_price if average_price is not None else "Not Available"
        except:
            specs["Price (Lakh INR)"] = "Not Available"
        
        specifications.append(specs)
    
    except Exception as e:
        print(f"Serial ID: {serial_id} - Error fetching specs for {truck_name}")

# Save the specifications to a CSV
df_specs = pd.DataFrame(specifications)
df_specs.to_csv('truck_specifications_with_prices.csv', index=False)

# Close the browser session
driver.quit()

print("Truck specifications and prices fetched and saved successfully!")
```

Cleaning Data

```python
import pandas as pd
import re

# Clean up the column 'Performance - Max Power' to convert all entries to kW
def clean_max_power(value):
    if pd.isna(value):
        return value
    value = value.lower()
    if 'kw' in value:
        return float(re.findall(r'\d+\.?\d*', value)[0])
    elif 'hp' in value:
        return round(float(re.findall(r'\d+\.?\d*', value)[0]) * 0.7457, 2)  # Convert hp to kW
    return None

# Clean up the column 'Performance - Max Torque' to remove 'Nm'
def clean_max_torque(value):
    if pd.isna(value):
        return value
    return float(re.findall(r'\d+\.?\d*', value)[0])

# Clean up the column 'Performance - Gradeability (%)' to ensure numerical values
def clean_gradeability(value):
    if pd.isna(value):
        return value
    try:
        return float(re.findall(r'\d+\.?\d*', value)[0])
    except:
        return None  # For non-numeric values, replace with NaN

# Clean up the column 'Performance - Range' to handle 'X-Y' type entries
def clean_range(value):
    if pd.isna(value):
        return value
    if '-' in value:
        values = [float(v) for v in re.findall(r'\d+\.?\d*', value)]
        return sum(values) / len(values)  # Take the average of the range
    return float(re.findall(r'\d+\.?\d*', value)[0])

# Clean up 'Performance - Battery Capacity' for multiple unit types (kWh, Ah)
def clean_battery_capacity(value):
    if pd.isna(value):
        return value
    value = value.lower()
    if 'kwh' in value:
        return float(re.findall(r'\d+\.?\d*', value)[0])
    elif 'ah' in value:
        return round(float(re.findall(r'\d+\.?\d*', value)[0]) / 1000, 2)  # Convert Ah to kWh approx
    elif '&' in value:
        values = [float(v) for v in re.findall(r'\d+\.?\d*', value)]
        return sum(values)  # Add both kWh values
    return None

# Clean up the 'Transmission & Loading Capacity - GVW / GCW (Kgs)' to remove 'kg' units
def clean_gvw_gcw(value):
    if pd.isna(value):
        return value
    return float(re.findall(r'\d+\.?\d*', value)[0])

# Clean up 'Performance - Max Speed (km/h)' to convert any range to an average value
def clean_max_speed(value):
    if pd.isna(value):
        return value
    if '-' in value:
        values = [float(v) for v in re.findall(r'\d+\.?\d*', value)]
        return sum(values) / len(values)  # Take average of the range
    return float(re.findall(r'\d+\.?\d*', value)[0])

# Clean 'Price (Lakh INR)' to replace 'Not Available' with 3
def clean_price(value):
    if value == 'Not Available':
        return 3
    return value

# Apply cleaning functions to the corresponding columns
df['Performance - Max Power'] = df['Performance - Max Power'].apply(clean_max_power)
df['Performance - Max Torque'] = df['Performance - Max Torque'].apply(clean_max_torque)
df['Performance - Gradeability (%)'] = df['Performance - Gradeability (%)'].apply(clean_gradeability)
df['Performance - Max Speed (km/h)'] = df['Performance - Max Speed (km/h)'].apply(clean_max_speed)
df['Performance - Range'] = df['Performance - Range'].apply(clean_range)
df['Performance - Battery Capacity'] = df['Performance - Battery Capacity'].apply(clean_battery_capacity)
df['Transmission & Loading Capacity - GVW / GCW (Kgs)'] = df['Transmission & Loading Capacity - GVW / GCW (Kgs)'].apply(clean_gvw_gcw)
df['Price (Lakh INR)'] = df['Price (Lakh INR)'].apply(clean_price)

# Check the cleaned data
print("Cleaned DataFrame structure:")
print(df.info())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
import re

# Step 1: Set up ChromeOptions to disable/handle permissions
chrome_options = Options()
chrome_options.add_argument("--disable-notifications")
chrome_options.add_argument("--disable-geolocation")

# Install and set up ChromeDriver using WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# Set an explicit wait to handle dynamic elements
wait = WebDriverWait(driver, 20)

# List of tab links and their respective IDs to target for extracting specifications
tabs = {
    "performance": "//a[@id='nav-performance']",
    "dimensions": "//a[@id='nav-dimensions']",
    "brakes_suspension": "//a[@id='nav-brakes_suspension']",
    "clutch_transmission": "//a[@id='nav-clutch_transmission']",
    "body_cabin": "//a[@id='nav-body_cabin']",
    "tyre": "//a[@id='nav-tyre']",
    "charging": "//a[@id='nav-charging']"
}

# Load the rickshaw links from the previously saved CSV file (or list)
rickshaw_links = pd.read_csv('e_rickshaw_links.csv')['Link'].dropna().tolist()

# Function to extract the rickshaw name from the link
def extract_rickshaw_name(link):
    # Everything after 'en/' is the name, replacing '/' with spaces
    name_part = link.split('/en/')[-1].replace('/', ' ').strip()
    return name_part

# Function to extract price and calculate average if there is a range
def extract_and_process_price(driver):
    try:
        # Locate the price element by its class name
        price_element = driver.find_element(By.XPATH, "//p[contains(@class,'newTruckSingle-price')]")
        price_text = price_element.text.strip()
        
        # Extract the price range or single price using regular expressions
        price_numbers = re.findall(r'[\d.]+', price_text)  # Extract all numbers
        if len(price_numbers) == 2:
            price_low = float(price_numbers[0])
            price_high = float(price_numbers[1])
            average_price = (price_low + price_high) / 2
        elif len(price_numbers) == 1:
            average_price = float(price_numbers[0])
        else:
            average_price = None
        
        return average_price
    except Exception as e:
        print(f"Error extracting price: {e}")
        return None

# Step 2: Function to extract specifications based on the tab links
def extract_specifications(link):
    rickshaw_name = extract_rickshaw_name(link)  # Extract the name from the link
    print(f"\nAttempting to visit: {link} ({rickshaw_name})")
    driver.get(link)
    
    # Wait for the page to load fully
    time.sleep(5)

    specs = {"Rickshaw Name": rickshaw_name}  # Add the rickshaw name to the specs

    try:
        print("Scrolling to the specification section...")
        spec_section = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'tab-content')))
        driver.execute_script("arguments[0].scrollIntoView(true);", spec_section)
        time.sleep(2)
    except Exception as e:
        print(f"Error locating the spec section: {e}")
        return specs

    # Extract the price and add to the specs
    average_price = extract_and_process_price(driver)
    if average_price is not None:
        specs['Average Price (Lakh INR)'] = average_price
    else:
        specs['Average Price (Lakh INR)'] = 'Price Not Available'

    # Loop through the tabs and extract the relevant table data after clicking the tab
    for tab_name, tab_xpath in tabs.items():
        try:
            print(f"Clicking tab: {tab_name}...")
            # Click the tab to make its content visible
            tab_link = driver.find_element(By.XPATH, tab_xpath)
            driver.execute_script("arguments[0].click();", tab_link)
            time.sleep(2)  # Allow time for the content to load after the click
            
            # Locate the corresponding tab-pane by ID and extract table data
            tab_pane = driver.find_element(By.ID, tab_name)
            rows = tab_pane.find_elements(By.XPATH, ".//table/tbody/tr")
            print(f"Number of rows found in tab {tab_name}: {len(rows)}")

            # Loop through each row and extract the key-value pairs
            for row in rows:
                try:
                    key = row.find_element(By.XPATH, "./td[1]").text.strip()
                    value = row.find_element(By.XPATH, "./td[2]").text.strip()
                    specs[key] = value
                    print(f"Found spec: {key}: {value}")
                except Exception as e:
                    print(f"Error extracting row: {e}")
                    continue

        except Exception as e:
            print(f"Error locating or processing tab: {tab_name}, Error: {e}")
    
    return specs

# Step 3: Loop through each rickshaw link and collect specifications
rickshaw_specs = []

for link in rickshaw_links:
    specs = extract_specifications(link)
    specs['Link'] = link  # Store the link with the specs for reference
    rickshaw_specs.append(specs)

# Step 4: Save the collected specifications to a CSV file
df_TractorJunction = pd.DataFrame(rickshaw_specs)
#df_TractorJunction.to_csv(r"C:\Users\risha\Downloads\EV_Analysis\Scripts\Assignment-Part2\Data\Raw Data\df_TractorJunction.csv", index=False)

# Close the browser session
driver.quit()

print("Rickshaw specifications fetched and saved successfully!")
```

```python
import requests
import pandas as pd

# Base API URL for fetching specifications
base_url = "https://apis.91trucks.com/v1/model/feature"

# Initialize a list to store the raw JSON responses
all_model_specs = []

# Function to extract brand and model from the Specification Link
def extract_brand_model_from_link(link):
    parts = link.split('/')  # Split the link by '/'
    brand = parts[-2]  # The second to last part is the brand
    model = parts[-1].replace('-', ' ').title()  # The last part is the model, format it properly
    return brand, model

# Function to fetch the specifications without using a variant ID
def fetch_specifications_without_variant(brand, model):
    params = {
        "categorySlug": "auto-rickshaws",
        "brandSlug": brand,
        "modelSlug": model.replace(" ", "-").lower(),
        "langCode": "en"
    }
    
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        return response.json()  # Return the parsed JSON response
    else:
        print(f"Failed to fetch data for {brand} {model}. Status Code: {response.status_code}")
        return None

# Load the CSV with Rickshaw links
# Assuming you have the CSV saved as 'dflinks.csv'
#dflinks = pd.read_csv('collected_rickshaw_links.csv')  # Load the CSV with Rickshaw links
```
